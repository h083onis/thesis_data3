```python
import tokenize
from io import BytesIO
from io import StringIO
```

```python
import subprocess

def format_code(file_path):
    try:
        # blackコマンドを呼び出してファイルを整形
        subprocess.run(["black", file_path])
        print(f"Code in {file_path} formatted successfully.")
    except Exception as e:
        print(f"Error formatting code: {e}")

# ファイルのパスを指定して整形
format_code("test2_format.py")
```

```python
with open('test.py', 'r', encoding='utf-8') as f:
    codes = f.read()
```

```python
out = ""
last_lineno = -1
last_col = 0
for token in tokenize.generate_tokens(StringIO(codes).readline):
    print(token)
    token_type = token[0]
    token_string = token[1]
    start_line, start_col = token[2]
    end_line, end_col = token[3]
    if start_line > last_lineno:
            last_col = 0
    if start_col > last_col:
        out += (" " * (start_col - last_col))
    # print(token)
    last_col = end_col
    last_lineno = end_line
    out += token_string
    if token_type == tokenize.OP and token_string == ':':
        out += '\n'
    
line_list = [line for line in out.split('\n') if line.strip() != '']
out = '\n'.join(line_list)
print(out)
    
```
