```python
import argparse
def read_args():
    parser = argparse.ArgumentParser()
    parser.add_argument('-json_file', type=str)
    parser.add_argument('-model_type', type=str, default='unigram')
    parser.add_argument('-model_prefix', type=str, default='openstack_spm_10000')
    parser.add_argument('-vocab_size', type=int, default=10000)
    parser.add_argument('-character_coverage', type=int, default=1.0)
    parser.add_argument('-user_defined_symbols', type=list, default=['<python>','<sep>'])
    parser.add_argument('-pad_id', type=int, default=-1)
    parser.add_argument('-type', type=str, default='code')
    return parser
params = read_args().parse_args([])
params
```

```python
import json
op_path = '../resource/preprocess_data/openstack.json'
dj_path = '../resource/preprocess_data/django2.json'
ansible_path = '../resource/preprocess_data/ansible2.json'
libcloud_path = '../resource/preprocess_data/libcloud2.json'
with open(op_path, 'r', encoding='utf-8') as f_op, \
    open(dj_path, 'r', encoding='utf-8') as f_dj,\
        open(ansible_path, 'r', encoding='utf-8') as f_ansible,\
        open(libcloud_path, 'r', encoding='utf-8') as f_libcloud: 
    op_data = json.load(f_op)
    dj_data = json.load(f_dj)
    ansible_data = json.load(f_ansible)
    libcloud_data = json.load(f_libcloud)
```

```python
ext_list = []
file_count = []
for commit in op_data:
    file_count.append(len(commit['codes']))
    for file in commit['codes']:
        ext = file['filepath'].split('.')[-1].lower()
        if ext not in ext_list:
            ext_list.append(ext)
for commit in dj_data:
    file_count.append(len(commit['codes']))
    for file in commit['codes']:
        ext = file['filepath'].split('.')[-1].lower()
        if ext not in ext_list:
            ext_list.append(ext)
for commit in ansible_data:
    file_count.append(len(commit['codes']))
    for file in commit['codes']:
        ext = file['filepath'].split('.')[-1].lower()
        if ext not in ext_list:
            ext_list.append(ext)
            print(commit['commit_id'])
for commit in libcloud_data:
    file_count.append(len(commit['codes']))
    for file in commit['codes']:
        ext = file['filepath'].split('.')[-1].lower()
        if ext not in ext_list:
            ext_list.append(ext)
            print(commit['commit_id'])
ext_list
```

```python
dj_data2 = []
for i, commit in enumerate(dj_data):
    if commit['codes'] == []:
        continue
    dj_dict = {}
    dj_dict['commit_id'] = commit['commit_id']
    dj_dict['timestamp'] = commit['timestamp']
    dj_dict['codes'] = []
    if len(dj_data2) < 18000:
        dj_dict['kind'] = 'train'
    else:
        dj_dict['kind'] = 'test'
    for file in commit['codes']:
        file_list = []
        for key, item in file.items(): 
            if key != 'filepath':
                if item != []:
                    lines = [dp.process(line) for line in item]
                    source = '<python> ' + ' <sep> '.join(lines)
                    if len(dj_data2) < 18000:
                        lines_list.append(source)
                    file_list.append(source)
                else:
                    file_list.append('')
        dj_dict['codes'].append(file_list)
    dj_data2.append(dj_dict)
        
        
```

```python
op_data2 = []
import pandas as pd
df = pd.read_csv('../resource/openstack.csv', index_col=0, header=0)
df.set_index('commit_id', inplace=True)
    
for commit in op_data:
    if commit['codes'] == []:
        continue
    op_dict = {}
    op_dict['commit_id'] = commit['commit_id']
    op_dict['timestamp'] = commit['timestamp']
    op_dict['codes'] = []
    if df.at[commit['commit_id'], 'strata'] != df['strata'].max():
        op_dict['kind'] = 'train'
    else:
        op_dict['kind'] = 'test'
    for file in commit['codes']:
        file_list = []
        for key, item in file.items(): 
            if key != 'filepath':
                if item != []:
                    lines = [dp.process(line) for line in item]
                    source = '<python> ' + ' <sep> '.join(lines)
                    if df.at[commit['commit_id'], 'strata'] != df['strata'].max():
                        lines_list.append(source)
                    file_list.append(source)
                else:
                    file_list.append('')
        op_dict['codes'].append(file_list)
    op_data2.append(op_dict)
```
